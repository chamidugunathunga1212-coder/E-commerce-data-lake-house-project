# Load the bronze layer tables

```python
CATALOG = "e-commerce-lake-house-project"
SCHEMA = "bronze"

def load_table(table_name):
    return spark.read.table(f"`{CATALOG}`.{SCHEMA}.{table_name}")

df_orders      = load_table(BRONZE_TABLES['orders'])
df_customers   = load_table(BRONZE_TABLES['customers'])
df_items       = load_table(BRONZE_TABLES['order_items'])
df_products    = load_table(BRONZE_TABLES['products'])
df_payments    = load_table(BRONZE_TABLES['payments'])
df_reviews     = load_table(BRONZE_TABLES['reviews'])
df_sellers     = load_table(BRONZE_TABLES['sellers'])
df_translation = load_table(BRONZE_TABLES['translation'])
```

# Data Profiling

##  Row counts & column counts

```python
tables = {
    "orders":      df_orders,
    "customers":   df_customers,
    "order_items": df_items,
    "products":    df_products,
    "payments":    df_payments,
    "reviews":     df_reviews,
    "sellers":     df_sellers,
    "translation": df_translation,
}

print("Dataset Overview")
print("=" * 45)
print(f"{'Table':<18} {'Rows':>10}  {'Columns':>8}")
print("-" * 45)

for name, df in tables.items():
    row_count = df.count()   
    col_count = len(df.columns)
    print(f"{name:<18} {row_count:>10,}  {col_count:>8}")
```
